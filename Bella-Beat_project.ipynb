change the datatype of the ActivityDate col

```python
df['ActivityDate']=pd.to_datetime(df['ActivityDate'])
df.info()
df['ActivityDate'].head(5)
```

Creating new column by separating the date into day of the week for further analysis.

```python
df['Days_ofWeek']=df['ActivityDate'].dt.day_name()
df['Days_ofWeek'].head(5)
```

Creating new column total_mins being the sum of total time logged.



```python
df['Total_Mins']=df['VeryActiveMinutes']+df['FairlyActiveMinutes']+df['LightlyActiveMinutes']+df['SedentaryMinutes']
df['Total_Mins'].head(5)
```

Creating new column by converting total_mins to number of hours.

```python
df['Total_hours']=round(df['Total_Mins']/60)
df['Total_hours'].head(5)
```

share the insight & visualizations 

```python
plt.style.use("default")
plt.figure(figsize=(5,5))
plt.hist(df['Days_ofWeek'],bins = 7,width = 0.4, color = "lightskyblue", edgecolor = "black")
plt.xticks(rotation=45)
plt.xlabel("days of the week")
plt.ylabel("Frequency")
plt.title("No. of times users logged in app across the week")
plt.grid(True)
plt.show()
```

relation between the total foot steps and Calories 

```python
median_calories = 2303
median_steps = 7637

plt.figure(figsize=(5,5))
plt.scatter(x=df['TotalSteps'],y=df['Calories'],c=df['Calories'],cmap = "Spectral")
plt.colorbar(orientation = "vertical")
plt.axvline(median_steps, color = "Blue", label = "Median steps")
plt.axhline(median_calories, color = "Red", label = "Median calories burned")
plt.xlabel('total-steps')
plt.ylabel('total-calories')
plt.grid(True)
plt.title("Calories burned for every step taken")
plt.legend()
plt.show()
```

Relation between total of hours and calories

```python
median_calories = 2303
median_hours = 20
median_sedentary = 991 / 60
plt.scatter(x=df['Total_hours'],y=df['Calories'],c=df['Calories'],cmap = "Spectral")
plt.colorbar(orientation = "vertical")
plt.xlabel('hours')
plt.ylabel('calories')
plt.axhline(median_calories,color='blue',label='median-calories')
plt.axhline(median_sedentary,color='blue',label='median_sedentary')
plt.axvline(median_hours,color='red',label='median_hours')
plt.grid(True)
plt.legend()
plt.title('Relation between total of hours and calories')
plt.show()
```

Percentage of Activity in Minutes


```python
very_active=df['VeryActiveMinutes'].sum()
Fairly_Active=df['FairlyActiveMinutes'].sum()
Lightly_Active=df['LightlyActiveMinutes'].sum()
Sedentary=df['SedentaryMinutes'].sum()
mins_slices =[very_active,Fairly_Active,Lightly_Active,Sedentary]
labels=['very_active','Fairly_Active','Lightly_Active','Sedentary']
mins_explode=[0,0,0,0.1]
colours = ["lightcoral", "yellowgreen", "lightskyblue", "darkorange"]
plt.pie(mins_slices,labels=labels,colors=colours,autopct='%1.1f%%',explode=mins_explode)
plt.title("Percentage of Activity in Minutes")
plt.tight_layout()
plt.show()

```
